# **METODOS DE LISTA**

```python
# tamanho de uma lista

clubs = ["Barcelona", "Chelsea", "PSG", "GALO", "Real", "Brasil", "Lazio"]

print(len(clubs))
```

```python
#inserindo dados em lista

clubs.append("Liverpool")
print(clubs)
```

```python
#inserindo dados em uma posição especifica

clubs.insert(0, "Man United")
print(clubs)
```

```python
# juntando listas

stadiuns = ["old trafford", "Camp Nou", "Stamford Bridge", "Allianz Arena"]

clubs.extend(stadiuns)
print(clubs)
```

```python
# rmovendo dados de uma lista

paises = ["brasil", "portugal", "argentina", "espanha"]

paises.remove("argentina")
print(paises)

paises.pop() #remove o ultimo elemento, ou o que eu desejar
print(paises)


```

```python
# metodo del

paises = ["brasil", "portugal", "argentina", "espanha"]

del paises[0]
print(paises)

#para limpar a lista e o clear
paises.clear()
print(paises)
```
