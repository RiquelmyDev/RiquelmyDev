# **OPERADORES LOGICOS**

são utilizados para conectar duas ou mais expressões boleanas, ou seja, expressões que retornam true ou false. o resultado final será um valor booleano

```python
#operador E (and) (as duas expressões precisam ser verdade, para ser True)

idade = 25
tem_carteira = True

idade >= 18 and tem_carteira
```

```python
from logging import fatal
# operador OU, precisa que uma das expressões seja verdadeira,
#para todo resto tambem ser verdade
tem_ingresso = False
nome_na_lista = True

tem_ingresso or nome_na_lista
```

```python
#OPERADOR NAO

logado =  False

not logado
```

```python
#PRIORIDADE DOS OPERADORES

"""
primeiro vai ser executado oque estiver em = ()
se não:
1 not
2 and
3 or
"""
```
