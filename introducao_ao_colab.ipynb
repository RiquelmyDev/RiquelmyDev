# **LAÇOS DE REPETIÇÃO FOR**

FOR É UMA ESTRUTURA DE REPETIÇÃO USADA PARA PERCORRER UMA SEQUENCIA DE DADOS COMO LISTAS, STRINGS INTERVALOS NUMERICOS, ENTRE OUTROS

```python
# UTILIZANDO FOR COM STRINGS

for x in "Python para dados":
    print(x)
```

```python
# EXEMPLO EM LISTAS

nomes = ["Joana", "Maria", "Josefa", "Mariana", "Geralda"]

for i in nomes:
    print(i)
```

```python
# for para numericos (for in range) regras a baixo:

"""
for i in range(inicio, fim, passo)

1) for i in range (fim) - declara que inicia em 0, e o valor é o fim
2) for i in range (inicio, fim) - declara que inicia em inicio, e o valor é o fim
3) for i in range (inicio, fim, passo) - declara que inicia em inicio, e o valor é o fim, e o passo

"""
```

```python
for i in range(11):
    print(i)
```

```python
for x in range (2, 21):
    print(x)
```

```python
for j in range(0, 22, 3):
    print(j)
```

```python
for i in range(1,10):
    print(i)
else:
    print("Fim da execução do for")
```

# **WHile**

é uma estrutura de um laço de repetição, que continua executando um bloco de codigo enquanto a condição definifa for verdadeira

assim que a condição se torna falsa, O LOOP para automaticamente.

```python
#Sintaxe while

"""
while condição: (enquanto for verdadeiro)
    bloco de codigo
"""
```

```python
i = 1

while i <= 10:
    print("contagem: ", i)
    i += 1
```

```python
senha = input("Digite sua senha: ")

while senha != "python":
    senha = input("Senha incorreta! Tente novamente: ")
print("Senha correta! Acesso liberado!!")
```

```python
numero = int(input("Digite um numero entre 1 e 10: "))

while numero < 1 or numero > 10:
    int(input ("Numero invalido, tente novamente: "))
print(numero)
```

# **WHILE TRUE**

```python

password = input("Enter your password: ")

while True:
    confirm_password = input("Confirm your password: ")

    if confirm_password == password:
        print("Password is okay, register!")
        break
    else:
        print("Passwords do not match, try again.")


```

```python
i = 1

while i < 10:
    print(i)
    if i == 3:
      break
    i += 1
```

```python
# continue (com exceção de: )

x = 0

while x < 21:
    x += 1
    if x == 10:
        continue
    else:
        print(x)
```
